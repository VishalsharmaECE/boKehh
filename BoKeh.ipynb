```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
```

```python
from bokeh.plotting import figure, output_file, show
from bokeh.sampledata.iris import flowers

output_file('test.html')

p = figure(title = 'test_flower')
p.xaxis.axis_label= 'Petal length'
p.yaxis.axis_label= 'Pteal width'
p.circle(x = flowers['petal_length'], y = flowers['petal_width'])
show(p)
```

```python
from bokeh.plotting import figure, output_file, show
from bokeh.sampledata.iris import flowers

output_file('line.html')

p = figure(title = 'line_graph')
p.xaxis.axis_label= ' x axis '
p.yaxis.axis_label= ' y axis'
p.line(x = flowers['petal_length'], y = flowers['petal_width'])
show(p)
```

```python
x = [1,2,3,4]
y = [4,5,6,7]

output_file('lines.html')

p = figure(title = 'line plot')
p.line(x,y)
show(p)
```

```python
x = [1,2,3,4]
y = [4,5,6,7]

output_file('lines.html')

p = figure(title = 'line plot')
p.scatter(x,y, fill_color = 'red', legend_label = 'red')
show(p)
```
